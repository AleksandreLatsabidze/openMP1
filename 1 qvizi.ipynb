```python
%%writefile parallel_for_demo3.cpp
#include <iostream>
#include <omp.h>
using namespace std;
int main()
{
  const int SIZE = 10;
  int A[SIZE], B[SIZE], C[SIZE];

  for(int i = 0; i < SIZE; i++)
  {
    A[i] = i;
    B[i] = i * 2;
  }

  #pragma omp parallel for
  for(int i = 0; i < SIZE; i++){
    C[i] = A[i] + B[i];

    // ჩვენ ვნახავთ რომელი თრედი რომელ იტერაციას ასრულებს
    cout << "თრედი " << omp_get_thread_num() << " ასრულებს i= " << i << " შედეგი C[i] = " << C[i] << endl;
  }

  for(int i = 0; i < SIZE; i++){
    cout << C[i] << endl;
  }
  return 0;

}
```

pragma omp parallel for ქმნის რამდენიმე თრედს(ნაკადს) და რამდენიმე ნაკადი ინაწლიებს მასივის შეკრებას პარალელურად.

```python
%%writefile reduction_demo.cpp
#include <iostream>
#include <omp.h>
using namespace std;
int main()
{
  int sum = 0;
  omp_set_num_threads(4);

  #pragma omp parallel for reduction(+:sum)
  for(int i = 0; i <= 100; i++)
  {
    sum += i;
  }
  cout << "Sum from 1 to 100 = " << sum << endl;
  return 0;
}
```

OpenMP 4 თრედად იყოფს ციკლს 0–დან 100-მდე.
თითოეული თრედი აკეთებს საკუთარ ნაწილობრივ sumს.
reduction(+:sum) აერთიანებს ყველა ნაწილობრივ sumს ერთ საბოლოო ჯამში.
Race Condition აღარ ხდება, და საბოლოო ჯამი ყოველთვის არის 5050. ანუ ერთად ითვლიან თრედები და ჯამს და საბოლოოდ openMP თვითონ აერთიანებს უსაფთხოდ შედებბს

```python
%%writefile sum_timing.cpp
#include <iostream>
#include <omp.h>
using namespace std;

int main() {
    const int SIZE = 10000000; // 10 მილიონი ელემენტი
    int* arr = new int[SIZE];

    // მასივის დატვირთვა .
    for (int i = 0; i < SIZE; i++)
    {
      arr[i] = i + 1;
    }
    //სერიული ჯამი
    double t1 = omp_get_wtime();
    long long sum_serial = 0;
    for (int i = 0; i < SIZE; i++) sum_serial += arr[i];
    t1 = omp_get_wtime() - t1;
    cout << "Serial sum = " << sum_serial << ", time = " << t1 << " s" << endl;

    // პარალელური ჯამი reduction-ით
    long long sum_parallel = 0;
    double t2 = omp_get_wtime();

    #pragma omp parallel for reduction(+:sum_parallel)
    for (int i = 0; i < SIZE; i++) sum_parallel += arr[i];

    t2 = omp_get_wtime() - t2;
    cout << "Parallel sum = " << sum_parallel << ", time = " << t2 << " s" << endl;

    delete[] arr;
    return 0;
}
```

pragma omp parallel for
კომპილატორს ეუბნება: გამოიყენე რამდენიმე თრედი ციკლის გასაკეთებლად.
ციკლის იტერაციები ნაწილდება თრედებს შორის, რომ სამუშაო სწრაფად დასრულდეს.reduction(+:sum_parallel) sum_parallel ცვლადი იქნება ასევე პარალელურად უსაფრთხო.
თითოეული თრედი აკეთებს საკუთარ ნაწილობრივ ჯამს (საკუთარ sum_parallel-ს).
ციკლის ბოლოს, OpenMP ყველას აერთიანებს ერთ ჯამში.
Race Condition არ ხდება, შედეგი ყოველთვის სწორია. კოდი გვიჩვენებს, რომ პარალელური მეთოდი უფრო სწრაფია დიდი მასივისთვის, ხოლო სერიული მეთოდი მუშაობს ერთ თრედზე.
Race Condition არ ხდება, რადგან პარალელურ ჯამში გამოვიყენეთ reduction. omp_get_wtime() ეს ფუნქცია OpenMP-ს დროის გაზომვის ფუნქციაა.ანუ მარტივად რო გავიგოთ სერიული ჯამი ერთი თრედი ითვლის მთელი მასივის ჯამს, ამიტომ უფრო ნელია.
პარალელური ჯამი (reduction) რამდენიმე თრედი ერთად ითვლის მასივის ნაწილებს, ბოლოს OpenMP აერთიანებს ჯამებს, ამიტომ უფრო სწრაფია.
